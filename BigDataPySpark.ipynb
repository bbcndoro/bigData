```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("Latihan Tambah Kolom & Filter").getOrCreate()

# Buat data
data = [("Andi", 23, "Jakarta"), ("Budi", 30, "Bandung"),
        ("Cici", 27, "Surabaya"), ("Wawan", 32, "Semarang")]
df = spark.createDataFrame(data, ["nama", "umur", "kota"])

print("Data awal:")
df.show()

# 1. Tambah kolom umur + 5
df2 = df.withColumn("umur_plus_5", col("umur") + 5)
print("\nSetelah tambah kolom umur_plus_5:")
df2.show()

# 2. Filter umur > 25
df_filtered = df.filter(col("umur") > 25)
print("\nFilter umur > 25:")
df_filtered.show()

# Selesai
spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Buat SparkSession
spark = SparkSession.builder.appName("Latihan Filter & GroupBy").getOrCreate()

# Data awal
data = [
    ("Andi", 23, "Jakarta"),
    ("Budi", 30, "Bandung"),
    ("Cici", 27, "Surabaya"),
    ("Wawan", 32, "Semarang")
]

df = spark.createDataFrame(data, ["nama", "umur", "kota"])

# 1. Filter orang dengan umur > 25
print("Filter orang dengan umur > 25:")
df_filtered = df.filter(col("umur") > 25)
df_filtered.show()

# 2. Mengelompokkan dan Aggregasi: Hitung jumlah orang per kota
print("\nHitung jumlah orang per kota:")
df_grouped = df.groupBy("kota").count()
df_grouped.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, upper

# Buat SparkSession
spark = SparkSession.builder.appName("Latihan Upper & Sort").getOrCreate()

# Data awal
data = [
    ("Andi", 23, "Jakarta"),
    ("Budi", 30, "Bandung"),
    ("Cici", 27, "Surabaya"),
    ("Wawan", 32, "Semarang")
]

df = spark.createDataFrame(data, ["nama", "umur", "kota"])

print("Data awal:")
df.show()

# 1. Mengubah nilai kolom (Transformasi)
# Misalnya membuat nama menjadi huruf kapital semua
df_upper = df.withColumn("nama_upper", upper(col("nama")))
print("\nSetelah nama diubah jadi huruf kapital:")
df_upper.show()

# 2. Mengurutkan Data berdasarkan umur dari yang tertua
df_sorted = df.orderBy(col("umur").desc())
print("\nData diurutkan dari umur terbesar ke terkecil:")
df_sorted.show()
```
